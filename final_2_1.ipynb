- Товар – наименование товара
- Количество – число единиц купленного товара

Воспользуйтесь этими данными и выясните, какие пары товаров пользователи чаще всего покупают вместе. По сути, вам необходимо найти паттерны покупок, что позволит оптимизировать размещение продуктов в магазине, для удобства пользователей и увеличения выручки.

- 1_Товар – наименование первого товара
- 2_Товар – наименование второго товара
- Встречаемость – число раз, когда такая пара была встречена

Другими словами: 2 раза люди покупали одновременно чай и арбуз, 1 раз одновременно покупали арбуз и сгущёнку и 1 раз одновременно были куплены чай со сгущёнкой.

Напишите код на python для получения нужной таблицы и укажите 5 наиболее распространённых паттернов.


````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid", rc={'figure.figsize':(14,7)})
%matplotlib inline
````

````python
# читаем данные
df = pd.read_csv('https://stepik.org/media/attachments/lesson/409319/test1_completed.csv')
df.head(10)
````

````python
# смотрим размерность, проверяем полноту и типы данных
print("-- Shape:")
print(df.shape)
print("\n-- Null values:")
print(df.isnull().sum())
print("\n-- dtypes:")
print(df.dtypes)
print("\n-- nuinque:")
print(df.nunique())
````

````python
# для подсчета встречаемости (частотности) нам важен сам факт присутствия товара в корзине
# поэтому добавим колонку in_basket по которой и будем считать показтель встречаемости
df['in_basket'] = np.where(df['Количество'] > 0 ,1,0) 
df.head()
````

````python
# интуитивно хочется "повернуть" таблицу, т.е. сделать pivot
# причем можно повернуть так, чтобы строки были товарами, а колонки покупками
# или наоборот, строки будет покупками, а колонки товарами
df_pivot1 = df.pivot(index='id', columns='Товар', values='in_basket').fillna(0)
df_pivot1
````

````python
df_pivot2 = df.pivot(index='Товар', columns='id', values='in_basket').fillna(0)
df_pivot2
````

````python
# Начинаю гуглить способ построить матрицу частот по всем колонкам, нахожу такой вариант:
# https://stackoverflow.com/questions/56836996/how-can-i-create-frequency-matrix-using-all-columns
df_pairs = df_pivot1.T.dot(df_pivot1)
df_pairs
````

````python
# Цифры встречаемости в таблице выше хоть и похожи на нужные, но для уверенности их нужно перепроверить.
# Исходную табличку сгруппируем по покупкам и список товаров в каждой покупке преобразуем в строку.
# Так проверить "встречаемость" будет проще.
df_check = df[['id','Товар']] \
    .groupby('id', as_index=False) \
    .agg(lambda x: x.tolist())
df_check['Товар'] = df_check['Товар'].astype(str)
df_check.head()
````

````python
# Теперь проверим пару примеров
# "Абрикос вяленый" и "Алыча вяленая" = должно быть 6 раз
# "Абрикос вяленый" и "Ягоды Годжи" = должно быть 7 раз
сheck1 = len(df_check.query('Товар.str.contains("Абрикос вяленый") and Товар.str.contains("Алыча вяленая")'))
сheck2 = len(df_check.query('Товар.str.contains("Абрикос вяленый") and Товар.str.contains("Ягоды Годжи")'))
print("Абрикос вяленый и Алыча вяленая =", сheck1)
print("Абрикос вяленый и Ягоды Годжи =", сheck2)
````

````python
# пересечение одинаковых товаров нам не интересны, они наоборот только мешают
# поэтому диагональ заполним нулями
np.fill_diagonal(df_pairs.to_numpy(), 0)
df_pairs
````

````python
# Теперь от этой матрицы частот, нам нужно взять или левую нижнюю часть относительно диагонали или верхнюю правую,
# потому что они по сути друг друга дублируют
df_final = pd.DataFrame(df_pairs.mask(np.triu(np.ones(df_pairs.shape)).astype(bool)).stack())
df_final
````

````python
# сбрасываем индексы, переименовываем колонки, сортируем по встречаемости (по убыванию)
df_final.reset_index(1) \
    .rename(columns={'Товар':'2_Товар'}) \
    .reset_index() \
    .rename(columns={'Товар':'1_Товар'}) \
    .rename(columns={0:'Встречаемость'}) \
    .sort_values(by='Встречаемость', ascending=False).head(10)
````
